## VISUALISATIONS (Pie Chart + Line Plot)

```python
columns = ['Hour','Holiday','Weekdays','Weather-situation',]

plt.figure(figsize=(12,18))
for i, column in enumerate(columns,1):
    plt.subplot(3,3,i)
    data = ds1[column].value_counts()
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('husl',len(data)))
    plt.title(f'Pie Chart of {column}')
    
    
plt.tight_layout()
plt.show()
```

 PIE CHART
 
 Above I have used Pie Chart to represent the ratio of data insights.
 
1.Hour - An average 4 % ,each from 0 to 12 Hour has.

2.Holiday - As we can see, around 96% there is no holiday.

3.Weekdays - Clearly with the help of Pie chart it is showing that every weekdays shared an average same amount of ratio.

4.Weather-sit - Clear weather is around 60% and Mist is around 32%.

# VISUALISATIONS (Box Plot + Histogram Plot)

```python
columns = ['Humidity','Windspeed','Total Count']

plt.figure(figsize=(12,8))


for i,col in enumerate(columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=ds2[col], palette='Set2')
    plt.title(f'Box Plot for {col}')
    plt.xlabel(col)
    
plt.tight_layout()
plt.show()
```

 BOX PLOT

 Here, I have used Box Plot to represent the numerical data for the Humidity, Windspeed & Total Count.
 
 It is showing that there is some outliers in Windspeed & Totl Count.

 HISTOGRAM PLOT

```python
columns = ['Humidity','Windspeed','Total Count']

plt.figure(figsize=(15, 12))

for i, col in enumerate(columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(ds2[col], kde=True, bins=20)
    plt.title(f'Histogram Plot showing {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    
plt.tight_layout()
plt.show()
    
```

 This code snippet provides a clear and insightful visualization of the distribution of the three specified columns in above dataset. 
 
Each histogram plot shows the frequency distribution, and the KDE curve gives an idea of the underlying distribution shape.
